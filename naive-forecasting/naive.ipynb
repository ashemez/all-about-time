# It's all about time


```python
import pandas as pd
import yfinance as yf
```

NVDA stock price data.

```python
nvidia_ticker = yf.Ticker('NVDA')
nvidia_df = nvidia_ticker.history(start='2020-06-01', end='2022-06-01')
nvidia_df = nvidia_df['Close'].to_frame()
nvidia_df.head(5)
```

```python
nvidia_df.plot()
```

The simplest way of predicting the timeseries data is assuming the next value to be the same with the previous value. This is called `Naive prediction`.

```python
nvidia_df['Naive Prediction'] = nvidia_df.shift()
nvidia_df.head(5)
```

```python
nvidia_df.plot()
```

## Differences

Forecasting on a stationary timeseries data is more reliable. The data plotted above is not stationary so we need to manipulate the timeseries data.

```python
nvidia_df['Difference'] = nvidia_df['Close'] - nvidia_df['Close'].shift()
nvidia_df.plot()
```

So, differences of actual and shifted data series is giving us a quite stationary version of the data.
